```python
#missing values
df.isnull().sum()

#here, in cuisines, there are 9 missing values, rest features have no missing values
```

```python
#another way of finding missing values
#we take a temp variable called features which iterates through df.columns and then if that specific feature is null or sum is > 0, we get the specific features with missing values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# finding missing values using heatmaps
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
#use seaborn documentation to study more about cbar, cmap and yticklabels
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns     #shows the columns of zomato dataset with country code as a feature
```

```python
#used to combine two datasets... here, zomato csv file and country code excel file
#on shows on which feature we want to combine the two tables and how specifies which table you need to focus on... left join or right join
pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
# to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
#finding out with respect to a particular country, how many records are there
final_df.Country.value_counts()
```

```python
#all country names with respect to that specific records
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values   #all the records with respect to country names
```

```python
## Pie chart   #we want to see which country has the maximum online orders using pie chart
plt.pie(country_val, labels=country_names)
```

```python
## pie chart --- top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.3f%%')
```

OBSERVATION:
Zomato maximum records or transactions are from India. Second and third largest are from US and UK respectively

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()

#last column is names 0 so we try to change that... this column repesents number of records corresponding to aggregate rating
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

OBSERVATION:
1. About 2148 users have not rated at all
2. When ratings are between 4.5 to 4.9 --> Excellent
3. When ratings are between 4.0 to 4.4 --> Very Good
4. When ratings are between 3.5 to 3.9 --> Good
5. When ratings are between 2.5 to 3.4 --> Average
6. When ratings are between 1.8 to 2.4 --> Poor
7. Maximum people who rated have rated the experience as Average or Good

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)      ##to increase the size of the barplot... if not used, x axis figures will be merged together and difficult to understand
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
## using this enhancement feature for heatmap
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

# hue parameter take column name for colour encoding and palette maps the color according to our choice
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

OBSERVATION:
1. maximum users have rated between 3.0 and 3.5
2. not rated rating count is very high

```python
## count plot
sns.countplot(x="Rating color", data=ratings, palette=['white','red','orange','yellow','green','green'])

# count is the frequency of each rating color in the dataset
```
